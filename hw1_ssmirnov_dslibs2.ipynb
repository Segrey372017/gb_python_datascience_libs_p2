## Практическое задание к Уроку 1 "Введение в задачу классификации. Постановка задачи и подготовка данных".

### Задание 1

Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Максимизировать точность классификации (Precision) целесообразно в случаях когда необходимо минимизировать ложно-положительные результаты (когда цена "пропуска" очень высока).<br>
__Пример1__: Биометрический доступ в серверную комнату: в комнату должен попадать только соответствующих персонал. Ущерб, который может быть нанесен "случайным человеком" перевешивает риск пропуска не спервого раза соответствующего специалиста (например с неснятыми очками, устаревшей фотографией).<br>
__Пример2__: Сортировка вторсырья - по конвейеру двигается мусор, робот отбирает ценное сырье - пластик, ткань, дерево, металл. Если как ткань будет распознано несколько пластиковый пакетов, то при последующей переработке кипы выйдет из-строя оборудование клиента на текстильной фабрике у сортировчного завода возникнут убытки. Аналогично по пластику если в контейнер с пластиком попадет текстиль волокнми будет загрязнена вся партия пластика, попавшая в экструдер. Поэтому в пограничных случаях пусть робот бездействует и объект проедет дальше - на свалку или ручную сортировку.<br>

Максимизировать отклик (Recall) целесообразно когда надо извечь максимальное количество подходящих объектов из выборки (желательно все). Здесь цена ложно-отрицательного выбора слишком высока.<br>
__Пример1__: Завод выпускает поршни тормозной системы автомобиля (грузовых и пассажирских микроавтобусов), детали идут на сборочный ковейер. Есть редкий дефект (примерно 400 изделий на миллион) - отслоение хромового покрытия, из-за чего у автомобиля через несколько месяцев использования отказывают тормоза. В этом случае робот, который проверяет качество покрытия должен гарантировано выбирать все бракованные детали, даже в случае если как бракованные будет помечено значительное чило годных. В этой задаче еще свою роль играет дисбаланс классов (явление довольно редкое).<br>
__Пример2__: Крупная IT-компания набирает большой штат аналитиков. Для прохождения первого этапа кандидат должен решить тестовое задание и написать мотивационное письмо. У компании есть возможность собеседовать большое число соискателей и отсеять наиболее способных. Команда биг дата разрабатывает модель, осуществляющую автоматическую проверку заданий, которая должна допускать до следующего этапа максимальное число вовлеченных и мотивированных кандидатов, даже если среди отобранных  будут невовлеченные и малопродуктивные специалисты, пропуск ценного специалиста, который мог в этот раз "прохладно отнестись" к выполнению задания для компании хуже.

### Задание 2

Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

F-мера обладает большей гибкостью, за счет коэффициента $ \beta$, благодаря которому можно настраивать метрику в сторону наиболее эффективного решения задачи - точности или отклика. Взятие средней величины возьмет составляющие метрики с равными весами т.е. показатель _F1-score_.
